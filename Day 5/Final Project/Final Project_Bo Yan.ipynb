### Data Quality Issues and Data Cleaning

1. Identify data quality issues in googleplaystore_user_reviews and do data cleaning.

```python
# Find column with most missing values
for col in df_google_play_store_user_reviews:
    print(col, len(df_google_play_store_user_reviews[df_google_play_store_user_reviews[col].isnull()]))

# Drop unused columns
df_google_play_store_user_reviews = df_google_play_store_user_reviews.drop(["Translated_Review","Sentiment"], axis = 1)

# Drop any rows with NaN
df_google_play_store_user_reviews = df_google_play_store_user_reviews.dropna(axis = 0)

# Drop duplicated rows
df_google_play_store_user_reviews = df_google_play_store_user_reviews.drop_duplicates()

df_google_play_store_user_reviews
```

```python
# Reorganize df_google_play_store_user_reviews with unique list of App names.
df_new_google_play_store_user_reviews = pd.DataFrame(columns = ["App_name", "Sentiment_Polarity", "Sentiment_Subjectivity"]) 

label_list = df_google_play_store_user_reviews["App"].to_list()
unique_list = df_google_play_store_user_reviews["App"].unique()
    
for i in range(0,len(unique_list)):
    num = 0
    sum_Sentiment_Polarity = 0
    sum_Sentiment_Subjectivity = 0
    for j in range(0, len(label_list)):
        if label_list[j] == unique_list[i]:
            num += 1
            sum_Sentiment_Polarity += df_google_play_store_user_reviews["Sentiment_Polarity"].to_list()[j]
            sum_Sentiment_Subjectivity += df_google_play_store_user_reviews["Sentiment_Subjectivity"].to_list()[j]
    df_new_google_play_store_user_reviews = df_new_google_play_store_user_reviews.append({'App_name' : unique_list[i], 'Sentiment_Polarity' : sum_Sentiment_Polarity/num, 'Sentiment_Subjectivity' : sum_Sentiment_Subjectivity/num}, ignore_index = True)

df_new_google_play_store_user_reviews = df_new_google_play_store_user_reviews.drop_duplicates("App_name", keep = 'last')

print("Shape of the cleaned google play store user reviews data:\n", df_new_google_play_store_user_reviews.shape)
print("\nHead of the cleaned google play store reviews data:\n", df_new_google_play_store_user_reviews.head())
```

2. Identify data quality issues in googleplaystore and do data cleaning.

```python
# Find column with most missing values
for col in df_google_play_store:
    print(col, len(df_google_play_store[df_google_play_store[col].isnull()])) 

# Drop duplicated rows
df_google_play_store = df_google_play_store.drop_duplicates("App", keep='last')

# Drop any rows with NaN
df_google_play_store = df_google_play_store.dropna(axis=0)

# Remove "+" and "," from the Installs attribute
df_google_play_store.Installs = df_google_play_store.Installs.str.replace('+','')
df_google_play_store.Installs = df_google_play_store.Installs.str.replace(',','')

# Convert Reviews and Installs to number.
df_google_play_store["Reviews"] = df_google_play_store["Reviews"].apply(pd.to_numeric, downcast='float', errors='coerce')
df_google_play_store["Installs"] = df_google_play_store["Installs"].apply(pd.to_numeric, downcast='float', errors='coerce')

# Column naming (convert column nemw to Snake_case)
df_google_play_store["Content_Rating"] = df_google_play_store["Content Rating"]
df_google_play_store["Last_Updated"] = df_google_play_store["Last Updated"]
df_google_play_store["Current_Ver"] = df_google_play_store["Current Ver"]
df_google_play_store["Android_Ver"] = df_google_play_store["Android Ver"]

print("\nShape of the cleaned google play store data:\n", df_google_play_store.shape)   
print("\nHead of the cleaned google play store data:\n", df_google_play_store.head())
```

### Variable Relationships

Perfect: If the value is near ± 1, then it said to be a perfect correlation: as one variable increases, the other variable tends to also increase (if positive) or decrease (if negative).

High degree: If the coefficient value lies between ± 0.50 and ± 1, then it is said to be a strong correlation.

Moderate degree: If the value lies between ± 0.30 and ± 0.49, then it is said to be a medium correlation.

Low degree: When the value lies below + .29, then it is said to be a small correlation.

No correlation: When the value is zero.

#### Question 2

2. What categories of Apps that a marketing strategy should targeted on the Google Play Store Apps?

```python
data_clustering = df_merge.copy()
data_clustering = data_clustering.drop(["App_name", "Category", "Size", "Type", "Price", "Genres", "Content_Rating", "Last_Updated", "Current_Ver", "Android_Ver"],1)
normalized = preprocessing.normalize(data_clustering)
print("The normalized data for clustering:\n", normalized)

X_input = StandardScaler().fit_transform(normalized)

# plot the elbow method
wcss = []
for i in range(1, 11):
    kmeans_p = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans_p.fit(X_input)
    wcss.append(kmeans_p.inertia_)
plt.figure(figsize = (15,8))
plt.subplot(2, 1, 1)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.grid()
plt.tight_layout()
plt.show()

# find the best elbow
kl = KneeLocator(range(1, 11), wcss, curve = "convex", direction = "decreasing")
elbow = kl.elbow
print("The best elbow (n_clusters) is", elbow, "\n")

# print the model, fit the model to the normalized data
kmeans = KMeans(n_clusters = elbow)
model = kmeans.fit(X_input)
print("model:\n", model, "\n")

# print the resulting cluster centers
centers = model.cluster_centers_
print("Cluster centers:\n", centers, "\n")

predict = kmeans.predict(X_input)

plt.figure(figsize = (15,8))
plt.subplot(2, 1, 2)
plt.scatter(X_input[:, 0], X_input[:, 1], c = predict, s = 50, cmap = 'viridis')
plt.scatter(centers[:, 0], centers[:, 1], c = 'black', s = 200, alpha = 0.5);
plt.grid()
```

```python
# Function that creates a DataFrame with a column for Cluster Number

def pd_centers(featuresUsed, centers):
	colNames = list(featuresUsed)
	colNames.append('Cluster')

	# Zip with a column called 'cluster' (index)
	Z = [np.append(A, index) for index, A in enumerate(centers)]

	# Convert to pandas data frame for plotting
	P = pd.DataFrame(Z, columns = colNames)
	P['Cluster'] = P['Cluster'].astype(int)
	return P
```

```python
features = ['Sentiment_Polarity', 'Sentiment_Subjectivity', 'Reviews', 'Installs', 'Rating']

P = pd_centers(features, centers)
print(P)
```

```python
# Function that creates Parallel Plots

def parallel_plot(data):
	my_colors = list(islice(cycle(['b', 'r', 'g', 'y', 'k']), None, len(data)))
	plt.figure(figsize = (15,8)).gca().axes.set_ylim([-20,+30])
	parallel_coordinates(data, 'Cluster', color = my_colors, marker = 'o')
```

3. Can we use NLP to train the Google Play Store dataset to predict reviews? 

```python
df = pd.read_csv("./googleplaystore_user_reviews.csv")

# Here I select the top 2000 items since it takes too long to run all items. 
df_user_reviews = df.head(2000)
```

```python
# Create a new dataframe for NLP analysis.
df_nlp = pd.concat([df_user_reviews.App, df_user_reviews.Translated_Review, df_user_reviews.Sentiment], axis = 1)
df_nlp.head()
```

```python
# Data cleaning
# Drop missing values
df_nlp.dropna(axis = 0,inplace = True)

# Drop duplicates rows
df_nlp.duplicated(subset = 'App').value_counts()

# Replace non-letter characters with space
first_text = df_nlp.Translated_Review[0]
text = re.sub("[^a-zA-Z]"," ",first_text)

# Convert upper case to lower case
text = text.lower()

# Review changes
print(df_nlp.Translated_Review[0]) 
print(text)
```

```python
df_nlp.Sentiment.value_counts()
```

```python
# Converting review types to int form 0=Positive, 1=Negative, 2=Neutral.
df_nlp.Sentiment = [0 if i == "Positive" else 1 if i == "Negative" else 2 for i in df_nlp.Sentiment]
df_nlp.head(10)
```

```python
# Remove irrelavent words
nltk.download("stopwords")
nltk.download("punkt")
```

```python
# Separate all words
text = nltk.word_tokenize(text)
text
```

```python
#lemmatization
nltk.download('wordnet')
```

```python
# Convert words to their root format.
wnl = nltk.WordNetLemmatizer()
text = [wnl.lemmatize(i) for i in text]
text = " ".join(text)
text
```

```python
text_list = []
for i in df_nlp.Translated_Review:
    text = re.sub("[^a-zA-Z]"," ",i)
    text = text.lower()
    text = nltk.word_tokenize(text)
    wnl = nltk.WordNetLemmatizer()
    text = [wnl.lemmatize(word) for word in text]
    text = " ".join(text)
    text_list.append(text)
    
text_list[:10]
```

```python
# Bag of words
max_features = 200000
count_vector = CountVectorizer(max_features = max_features, stop_words = "english")
sparce_matrix = count_vector.fit_transform(text_list).toarray()
all_words = count_vector.get_feature_names()
all_words[0:50]
```

```python
plt.subplots(figsize = (15,10))
wordcloud = WordCloud(background_color = "pink",width = 1024,height = 768).generate(" ".join(all_words[100:]))
plt.imshow(wordcloud)
plt.show()
```

```python
# Classification
y = df_nlp.iloc[:,1].values
x = sparce_matrix
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
#logistic Regression
logistic_regression = LogisticRegression()
logistic_regression.fit(x_train,y_train)
y_prediction = logistic_regression.predict(x_test)
cm = confusion_matrix(y_test, y_prediction) 
accuracy_logistic = logistic_regression.score(x_test,y_test)
print(cm)
print("\nlogistic_regression accuracy: ", accuracy_logistic)
```

```python
# Now Fitting Naive Bayes classifier to the Training set

classifier = GaussianNB()
classifier.fit(x_train, y_train)

# Predicting the Test set results
y_prediction = classifier.predict(x_test)

# Making the Confusion Matrix and find Accuracy
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_prediction)    
accuracy_naive_bays = accuracy_score(y_test, y_prediction)

print(cm)
print("\nNaive Bayes Classifier accuracy: ", accuracy_naive_bays)
```
