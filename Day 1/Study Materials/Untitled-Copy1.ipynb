```python
# Most common occupations

print(users[['user_id', 'occupation']].groupby('occupation').count().sort_values('user_id').tail(10))
student_filter = users['occupation']=='student'
print(len(users[student_filter]))
```

```python
len(users.zip_code.unique())
```

```python
# Zipcodes with max population

top_zipcode = users.zip_code.value_counts().head(10).index
print(top_zipcode)
```

```python
# Users in the most common Zipcodes
users[users['zip_code'].map(lambda x: x in top_zipcode)]
```

```python
# Age statistics
users['age'].describe()
```

```python
# Users within standard deviation of the mean
mean=users.age.mean()
std = users.age.std()
users_filtered=users[users.age.gt(mean-std) & users.age.lt(mean+std)]
```

```python
# Number of male students in the above range
filt2 = users_filtered['gender']=='M' 
filt3 = users_filtered['occupation']=='student'
users_filtered[filt2 & filt3]
```
